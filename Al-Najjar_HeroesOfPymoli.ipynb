```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
player_count = len(purchase_data["SN"].unique())
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# variables
unique_items = len(purchase_data["Item Name"].unique())

total_rev = purchase_data["Price"].sum()

number_of_purchases = len(purchase_data["SN"])

avg_price = total_rev / number_of_purchases

# Place all of the data found into a summary DataFrame
summary_table = pd.DataFrame({"Number of Unique Items": unique_items,
                              "Average Price": (f"${round(avg_price,2)}"),
                              "Number of Purchases": number_of_purchases,
                              "Total Revenue": (f"${total_rev}")}, index=[0])
summary_table
```

```python
# Using GroupBy in order to separate the data into fields according to "Gender" values
gender_count_df = purchase_data.groupby("Gender")["SN"].nunique()
gender_count_df.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
## create bins
bins = [-1, 9, 14, 19, 24, 29, 34, 39, 100]

# Create the names for the four bins
age_counts = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# add column by age
purchase_data["Age Counts"] = pd.cut(purchase_data["Age"], bins, labels=age_counts)

# Creating a group based off of the bins
purchase_data_age = purchase_data.groupby("Age Counts")

# variables
player_count_by_age = purchase_data_age["SN"].nunique()

player_percentage_by_age = (player_count_by_age / player_count) * 100
player_percentage_by_age = round(player_percentage_by_age,2)

# sumamrize in df
summary_age_df = pd.DataFrame({"Total Count": player_count_by_age,
                              "Percentage of Players": (f"{player_percentage_by_age})
summary_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# create variables
usernames = purchase_data.groupby('SN')

purchase_count_names = usernames['Price'].count()

total_purchase_names = usernames['Price'].sum()

avg_purchase_names = total_purchase_names / purchase_count_names
avg_purchase_names = round(avg_purchase_names,2)

# summarize in df
summary_names_df = pd.DataFrame({"Purchase Count": purchase_count_names,
                              "Average Purchase Price ($)": avg_purchase_names,
                                  "Total Purchase Value ($)": total_purchase_names})

# sort df for top spenders
top_spenders = summary_names_df.sort_values(["Total Purchase Value ($)"], ascending=False)
top_spenders
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# group
item_grouped = purchase_data.groupby(['Item ID', 'Item Name'])

# variables
purchase_count_item = item_grouped['Purchase ID'].count()

item_price = item_grouped['Price'].mean()

total_purchase_item = item_grouped['Price'].sum()

# summarize in df
summary_items_df = pd.DataFrame({"Purchase Count": purchase_count_item,
                              "Item Price ($)": item_price,
                                  "Total Purchase Value ($)": total_purchase_item})
summary_items_df

# sort df for purchase count
popular_items = summary_items_df.sort_values(["Purchase Count"], ascending=False)
popular_items
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# sort df for profit count
profitable_items = summary_items_df.sort_values(["Total Purchase Value ($)"], ascending=False)
profitable_items
```
